### Finding The outliers 

```python
numerical_vars = ['age', 'creatinine_phosphokinase', 'ejection_fraction', 
                  'platelets', 'serum_creatinine', 'serum_sodium', 'time']
def detect_outliers(df, features):
    outliers={}
    for var in features:
        Q1=df[var].quantile(0.25)
        Q3=df[var].quantile(0.75)
        IQR=Q3-Q1
        lb=Q1-1.5*IQR
        hb=Q3+1.5*IQR
        outliers_indices=df[(df[var]<lb) | (df[var]>hb)].index
        outliers[var]=outliers_indices
    return outliers

outliers=detect_outliers(df,numerical_vars)
print(outliers)

# Print only the outlier indices
for var, indices in outliers.items():
    print(f"Outliers in {var}:")
    print(indices)

# Plot outliers for each numerical variable
plt.figure(figsize=(12, 8))
for i, var in enumerate(numerical_vars, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=df[var])
    plt.title(f'Boxplot of {var} with Outliers')
    plt.xlabel(var)
plt.tight_layout()
plt.show()
```

### Fixing the outliers with its Mean Values 

```python
# Function to fix outliers by replacing them with mean values
def fix_outliers_with_mean(df, outliers):
    for var, indices in outliers.items():
        mean_val = df[var].mean()
        df.loc[indices, var] = mean_val
    return df

# Fix outliers with mean values
df_fixed = fix_outliers_with_mean(df, outliers)

# Print DataFrame to verify outliers are fixed
print("DataFrame after fixing outliers:")
print(df_fixed)
```

## Hypothesis Testing :
#### Aim: The goal of hypothesis testing is to determine whether there is sufficient evidence to reject the null hypothesis in favor of the alternative hypothesis, based on sample data and statistical analysis

#### Null Hypothesis: Is a statement that there is no significant difference, effect, or relationship between the variables in the population

#### Alternative Hypothesis(H1):It is a statement which asserts that there is a significant difference, effect, or relationship between variables in the population.

### T-test

```python
from scipy.stats import ttest_ind

# Define numerical variables for t-test (including 'age')
numerical_vars = ['age', 'creatinine_phosphokinase', 'ejection_fraction', 'platelets', 'serum_creatinine', 'serum_sodium', 'time']

# Perform t-test for each numerical variable
for var in numerical_vars:
    # Separate data for patients who survived and those who didn't
    survived = df[df['DEATH_EVENT'] == 0][var]
    not_survived = df[df['DEATH_EVENT'] == 1][var]

    # Perform t-test
    t_statistic, p_value = ttest_ind(survived, not_survived)

    # Print the results
    print(f"T-Statistic for {var}:", t_statistic)
    print(f"P-Value for {var}:", p_value)

    # Interpret the results
    alpha = 0.05  # significance level
    if p_value < alpha:
        print(f"Reject the null hypothesis. There is a significant difference in the average {var} between patients who survived and those who didn't.")
    else:
        print(f"Fail to reject the null hypothesis. There is no significant difference in the average {var} between patients who survived and those who didn't.")
    print("--------------------------------------------")
```

##### Based on the reulst a 'P' value is less than alpha value we can cleary reject the null hypothesis statement for some numerical variables which are 'age','ejection fraction','serun creatinine','serum sodium' and 'time' these have a significant impact on the patinet survival outcomes.Whereas in the other hand creatinine phosphokinase and platelet levels do not show a significant difference between survivors and non-survivors.

## Chi-Square Test

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load the dataset
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')

# Select categorical variables
categorical_vars = ['anaemia', 'high_blood_pressure', 'diabetes', 'sex', 'smoking']

# Perform chi-square test for each categorical variable
for var in categorical_vars:
    contingency_table = pd.crosstab(df[var], df['DEATH_EVENT'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    print(f"Chi-square test results for {var}:")
    print(f"Chi-square statistic: {chi2}")
    print(f"P-value: {p}")
    print(f"Degrees of freedom: {dof}")
    print("")

    if p < 0.05:
        print(f"Reject the null hypothesis. There is a significant association between {var} and Death Event.")
    else:
        print(f"Fail to reject the null hypothesis. There is no significant association between {var} and Death Event.")
    print("--------------------------------------------")
```

##### For the given datset  based on the chi-square test results, none of the categorical variables (anaemia, high blood pressure, diabetes, sex, smoking) appear to be significantly associated with patient survival status. Therefore, these factors may not be strong predictors of patient outcomes in heart failure cases.
